```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
computers=pd.read_csv('computers.csv')
```

```python
computers.info()
```

```python
computers.head()
```

```python
mean_value=computers['Minutes'].mean()
mean_value
```

```python
plt.scatter(computers['Units'],computers['Minutes'],color='blue')
plt.xlabel('Units')
plt.ylabel('Minutes')
```

Craete a models

```python
Model0=computers['Minutes'].mean()
Model1=10+12 * computers['Units']
Model2=6+18 * computers['Units']
Model3 = 4.16 + 15.50 * computers['Units']
```

```python
computers['Model0']=Model0
computers['Model1']=Model1
computers['Model2']=Model2
computers['Model3']=Model3
```

```python
computers
```

```python
fig,ax=plt.subplots()
ax.scatter(x="Units",y="Minutes",data=computers,label='actual repair time')
ax.plot(computers['Units'],computers['Model0'],color="red",label='Model0')
ax.plot(computers['Units'],computers['Model1'],color="green",label='Model1')
ax.plot(computers['Units'],computers['Model2'],color="black",label='Model2')
ax.plot(computers['Units'],computers['Model3'],color="yellow",label='Model3')
ax.set_ylabel("Minutes")
ax.set_xlabel("Units")
ax.set_title("Speculated Models")
ax.legend()
```

Build scikit learn model - linear regression

```python
X = computers[['Units']]
y = computers['Minutes']
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X,y)
```

```python
print("Intercept:",model.intercept_)
print("Coefficient:",model.coef_)
```

```python
Rsq = model.score(computers[["Units"]],y) * 100
Rsq
```
