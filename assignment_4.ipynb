1. Implement Naive Bayes classification.

Work with iris.csv. Find types of data. Accordingly to data, which Naive Bayes can implement here?  Perform  encoding with target columns. Implement suitable Naive Bayes classification.  Find out accuracy of model.



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB  
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("iris.csv")
```

```python
df
```

```python
df.info()
```

```python
le = LabelEncoder()
df['Species'] = le.fit_transform(df['Species'])
```

```python
X = df.drop('Species', axis=1)
Y = df['Species']
```

```python
X
Y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)
```

```python
nb_pred = nb_model.predict(X_test)
print("Naive Bayes Accuracy:", accuracy_score(y_test, nb_pred))
```

2. Implement SVM binary classification.

Work with iris.csv. Import necessary files to implement SVM binary classification. Prepare data for binary classification. Visualize data distribution using seaborn - pairplot. Implement binary classification.

```python
binary_df = df[df['Species'] != 2]
```

```python
X_bin = binary_df.drop('Species', axis=1)
y_bin = binary_df['Species']
```

```python
sns.pairplot(binary_df, hue="Species")
plt.show()
```

```python
X_train_b, X_test_b, y_train_b, y_test_b = train_test_split(X_bin, y_bin, test_size=0.2, random_state=42)
```

```python
svm_model = SVC(kernel='linear')
svm_model.fit(X_train_b, y_train_b)
```

```python
svm_pred = svm_model.predict(X_test_b)
print("SVM Binary Classification Accuracy:", accuracy_score(y_test_b, svm_pred))
```
